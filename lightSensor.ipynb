Needs to find the height of aluminum end caps, for which the device will sink in Isopar-K. 

The calculation were made under the assumptions, that our device consists of two solid aluminum cylindrical end cups and a hollow central (electronic) chamber made of polycarbonate. 

```python
import numpy as np

g = 9.8 * 100 # m/c^2 * 100 = cm/c^2

OD = 4*2.54 #cm
r_logger = OD/2 # cm
inner_diameter = 3 * 2.54 # cm
r_elChamber = inner_diameter/2 # cm radius of empty space inside electronic chamber for our hardware

wall_thickness = r_logger - r_elChamber # cm


rho_Al = 2.7 # g/cm^3
rho_IsoparK = 758.5 * 1000 / pow(100,3) # 758.5 kg/m * 1000 / pow(100,3) = g/cm
rho_polycarb = 1.22 # g/cm

h_elChamber_range = np.arange(10,30,1) #cm
h_al_range = np.arange(1,20,1) #cm

#h_elChamber = 20 #cm
#h_al = 5 #cm 

def V_cyllinder(r, h): 
    return np.pi * pow(r,2) * h 

def V_cyllinderWalls(r_cyllinder, r_internalParts, h):  
    return ( np.pi * pow(r_cyllinder,2) * h - np.pi * pow(r_internalParts,2) * h )

def F_gravity(rho, V):
    m = rho * V
    return m * g

def F_buoyant(V):
    return rho_IsoparK * g * V

for h_elChamber in range(20, 30, 1):
    for h_al in range(0, 15, 1):
        
        F_gravity_electronicChamber =  F_gravity( rho_polycarb, V_cyllinderWalls(r_logger, r_elChamber, h_elChamber) )

        F_gravity_theWholeLogger = F_gravity( rho_Al, V_cyllinder(r_logger, h_al) ) + F_gravity_electronicChamber

        F_upthrust_theWholeLogger = F_buoyant( np.pi * pow(r_logger,2) * ( h_al + h_elChamber ) ) 

        if (F_gravity_theWholeLogger > F_upthrust_theWholeLogger):
            print('Logger goes down (sinks in Isopar K)')
            print ('mass of aluminum end caps' , round( F_gravity( rho_Al, V_cyllinder(r_logger, h_al) ) / g ), 'gramm;', 
                                                                         ' height of aluminum end caps (both combined)' , h_al , ' cm')
            print ('mass of polycarbonate walls' , round( F_gravity( rho_polycarb, V_cyllinderWalls(5, 3.5, h_elChamber) ) / g ),
                                    'gramm ;',  ' height of electronic chamber ' , h_elChamber, 'cm\n'  )
            break 
```

Needs to find the distance that is needed for the ratchet to move, so that the distance between the optical targets changes from 10 to 17 cm

<img src="duga.png" style="height:300px">

```python
import numpy as np

#by using huygens-equation https://mathvox.com/huygens-equation/ 

#m1 = h1 / np.sin(alpha1)

#M1 = m1 * np.cos(alpha1)

#L = 2*m1 + (2m1 - M1)/3

#L = 2*m1 + (2m1 - m1 * np.cos(alpha1))/3

OD_logger = 5 #cm
maximum_width_between_two_optical_targets = 17 #cm
minimum_width_between_two_optical_targets = 10 #cm
distance_between_spring_fasteners = 22 #cm

M1 = distance_between_spring_fasteners/2
h1 = (maximum_width_between_two_optical_targets - OD_logger)/2
m1 = np.sqrt( pow(h1,2) + pow(M1,2) )
L = 2*m1 + (2*m1 - M1)/3
alpha1 = np.arcsin(h1/m1 )
```

```python
#L = 2*m2 + (2m2 - m2 * np.cos(alpha2))/3 
#m2 = h2 / np.sin(alpha2)
# L = 2*h2 / np.sin(alpha2) + ( 2*h2 / np.sin(alpha2) - h2 / np.sin(alpha2) * np.cos(alpha2) )/3

h2 = ( minimum_width_between_two_optical_targets - OD_logger )/2
import numpy as np
from sympy import *
from sympy.solvers import solveset
from sympy import Symbol
alpha2 = Symbol('alpha2')
solution = solve(L - 2*h2 / sin( alpha2 ) + (2*h2 / sin( alpha2 ) - h2 / sin( alpha2 ) * cos( alpha2 ))/3, alpha2)
alpha2 = np.float64( solution[0] )
m2 = h2 / np.sin(alpha2)
M2 = m2 * np.cos(alpha2)
print('ratchet needs to move on ' + str( "{:.1f}".format(M2*2 - M1*2) ) + ' cm, so that the distance between the optical targets changes from 10 to 17 cm '  )
```
